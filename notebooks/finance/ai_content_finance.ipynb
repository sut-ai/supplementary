<h2> Stationary Time Series </h2>


<div> 
    <font face="Arial" size=5 style="line-height:200%">
        <p>
            When we gather time series datasets from financial markets like Forex, Equities and stocks related data, stationarity is always the term that we come across. So the question is what is considered stationary data?
            Do we need to care about stationarity assumption and if we do what should we do?
        </p>
        <p> 
            Stationary data refers to the time series data that mean and variance do not vary across time. The data is considered non-stationary if there is a <strong>trend</strong> or <strong>seasonality or cycle(different from seasonability)</strong> observed in data. As shown in the picture bellow from here, only (b) and (g) are considered stationary. 
            <a href="https://towardsdatascience.com/stationarity-assumption-in-time-series-data-67ec93d0f2f#:~:text=What%20is%20stationary%20data%3F,seasonality%20observed%20from%20the%20data">Image source</a> 
        </p>
        <img src="./imgs/stationarity_example.png">
        <p> 
            In Financial Markets, the time series data is rarely stationary. 
            When we fit a stationary model to the time series data that we want to analyze,
            we should detect the stationarity of the data and remove the trend/seasonality effect from the data.
        </p>
        <h3> How do we know whether the data is stationary? </h3>
        <p> 
            <ul>
                <li style="margin-top:8px">
                      (NOT ROBUST)  — Observing the Chart
                        <p style="font-size: 18px; width:50%; margin-top:15px"> 
                            We can observe the chart and see if there is a trend or a seasonality in the data. 
                            If there is no trend or seasonality, then the data is stationary.
                        <a href="https://towardsdatascience.com/stationarity-assumption-in-time-series-data-67ec93d0f2f#:~:text=What%20is%20stationary%20data%3F,seasonality%20observed%20from%20the%20data"> For More Information </a>
                    </p>
                </li>
                <li style="margin-top:8px">
                    Statistical Approach — Augmented Dickey-Fuller test
                    <p style="font-size: 18px; width:50%; margin-top:15px"> 
                        In statistics and econometrics, an augmented Dickey–Fuller test (ADF) tests the null hypothesis that a unit root is present in a time series sample. The alternative hypothesis is different depending on which version of the test is used, but is usually stationarity or trend-stationarity.
                        <a href="https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test"> For More Information </a>
                    </p>
                </li>
            </ul>
        </p>
        <h3> What can we do if the data is non-stationary? </h3>
        <p> 
            <ul>
                <li style="margin-top:8px">
                    Remove Trend
                        <p style="font-size: 18px; width:50%; margin-top:15px"> 
                            If the data is non-stationary, we can remove the trend from the data. 
                            We can do this by fitting a linear regression model to the data and removing the trend from the data.
                            <a href="http://www.stat.yale.edu/Courses/1997-98/101/linreg.htm#:~:text=Linear%20regression%20attempts%20to%20model,to%20be%20a%20dependent%20variable."> For More Information </a>
                            Also there is a method called "First Difference" which is also called "Differencing" in finance.
                            It is a simple method to remove the trend from the data. But it is not robust.
                        </p>
                </li>
                <li style="margin-top:8px">
                    Remove Seasonality
                        <p style="font-size: 18px; width:50%; margin-top:15px"> 
                            If the data is non-stationary, we can remove the seasonality from the data. 
                            A simple way to correct for a seasonal component is to use differencing.
                            But this time we define the differencing factor equal to the time period of each season.
                            <a href="https://machinelearningmastery.com/time-series-seasonality-with-python/#:~:text=Removing%20Seasonality&text=The%20model%20of%20seasonality%20can,referred%20to%20as%20non%2Dstationary."> For More Information and More Robust Methods </a>
                        </p>
                </li>
                <li style="margin-top:8px">
                    Remove different volatility
                        <p style="font-size: 18px; width:50%; margin-top:15px"> 
                            We can remove different volatility from the data by rolling a window on data and calculating the standard deviation of the data.
                            Then normalize each data point by the standard deviation.
                        </p>
                </li>
            </ul>    
    </font>
</div>

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
### Methods 
def parser(s):
    return datetime.strptime(s, '%Y-%m')

def plot_series(series):
    plt.figure(figsize=(18,7))
    plt.plot(heater_series, color='red')
    plt.ylabel('Search Frequency for "Heater"', fontsize=16)

    for year in range(2004, 2021):
        plt.axvline(datetime(year,1,1), linestyle='--', color='k', alpha=0.5)

### Read Data
ice_cream_heater_df = pd.read_csv('data/ice_cream_vs_heater.csv', parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
ice_cream_heater_df = ice_cream_heater_df.asfreq(pd.infer_freq(ice_cream_heater_df.index))
heater_series = ice_cream_heater_df.heater

plot_series(heater_series)
```

```python
### Normalizing Data
avg, dev = heater_series.mean(), heater_series.std()
heater_series = (heater_series - avg) / dev

### First difference
heater_series = heater_series.diff().dropna()

### remove different volatility
annual_volatility = heater_series.groupby(heater_series.index.year).std()
heater_annual_vol = heater_series.index.map(lambda d: annual_volatility.loc[d.year])
heater_series = heater_series / heater_annual_vol

### remove seasonality
month_avgs = heater_series.groupby(heater_series.index.month).mean()
heater_month_avg = heater_series.index.map(lambda d: month_avgs.loc[d.month])
heater_series = heater_series - heater_month_avg

### plot
plot_series(heater_series)
plt.axhline(0, linestyle='--', color='k', alpha=0.3)
```

## Forecasting Financial data using Simple classic models 
### pair trading
<div style="width:90%; font-size:22px">
A pairs trade is a trading strategy that involves matching a long position with a short position in two stocks with a high correlation.
In this case, we are trading the pair of stocks, <b> MSFT and AAPL </b>
</div>


```python
import yfinance as yf
ticker_m = yf.Ticker("MSFT")
data_m = ticker_m.history(period = "10y")
data_m.head(2)
```

```python
high_low_m = data_m['High'] - data_m['Low']
high_close_m = np.abs(data_m['High'] - data_m['Close'].shift())
low_close_m = np.abs(data_m['Low'] - data_m['Close'].shift())

ranges_m = pd.concat([high_low_m, high_close_m, low_close_m], axis=1)
true_range_m = np.max(ranges_m, axis=1)
atr_m = true_range_m.rolling(30).sum()/30
ratr_m = atr_m / data_m.Close
```

```python
ticker_g = yf.Ticker("GOOG")
data_g = ticker_g.history(period = "10y")
data_g.head(2)
```

```python
high_low_g = data_g['High'] - data_g['Low']
high_close_g = np.abs(data_g['High'] - data_g['Close'].shift())
low_close_g = np.abs(data_g['Low'] - data_g['Close'].shift())

ranges_g = pd.concat([high_low_g, high_close_g, low_close_g], axis=1)
true_range_g = np.max(ranges_g, axis=1)

atr_g = true_range_g.rolling(30).sum()/30
ratr_g = atr_g / data_g.Close
```

```python
plt.figure(figsize=(18,5))
plt.plot(ratr_g, color="red", label="Google")
plt.plot(ratr_m, color="blue", label="MSFT")
plt.legend(loc="upper left")
```

```python
# Rate of relative change
rorc = (ratr_g - ratr_m)/np.abs(ratr_m)
rorc.dropna(inplace=True)

plt.figure(figsize=(18,5))
plt.plot(rorc, color="green")
```

### check stationarity

```python
from statsmodels.tsa.stattools import adfuller

def perform_adf_test(series):
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
```

```python
perform_adf_test(rorc)
```

So the p-value is less than 0.05, so we can reject the null hypothesis that the series is stationary.
