# Online Learning

In computer science, online machine learning is a method of machine learning in which data becomes available in a sequential order and is used to update the best predictor for future data at each step, as opposed to batch learning techniques which generate the best predictor by learning on the entire training data set at once. [This youtube tutorial](https://www.youtube.com/watch?v=O3gd6elZOlA) might be helpful to get more familiar with this family of ML methods 


## Continual learning
* Continual learning means constantly improving the learned model by processing continuous streams of information. Continual learning capabilities are essential for software systems and autonomous agents interacting in an ever changing real world.

* Ideally, what you want is a model that can learn from new examples in something close to real time. Not only predict in real time, but learn in real time, too.

# Semi-Suipervised Learning
Semi-supervised learning is an approach to machine learning that combines a small amount of labeled data with a large amount of unlabeled data during training. Semi-supervised learning falls between unsupervised learning (with no labeled training data) and supervised learning (with only labeled training data). It is a special instance of weak supervision.

* Pseudo labling: Assume that only a small portion of our training data is labled. To get the most out of our training data, we should assign correct lables to unlabled data. In pseudo labling, we traing a supervised model with our labled data, and utilize its predictions for assigning lables for our unlabled data. # Few-Shot Learning

Few-shot learning is the problem of making predictions based on a limited number of samples. Few-shot learning is different from standard supervised learning. The model should adapt efficiently with a small number of training data. In most cases we need to prepare a model that is capable of this fast adaptation. 


*   **Support Set**: Support set is meta learnings jargon. The small set of labeled images is called a support set. Note the differences between the training set and the support set. The training set is big. Every class in the training set has many samples. The training set is big enough for learning a deep neural network. In contrast, the support set is small. Every class has at most a few samples. In the training set, if every class has only one sample, it is impossible to train a deep neural network. Thus, in few-shot learning we have support sets instead of training sets.

*   **Query Set** Is the set of unlabled data that the model should predict after training of the support set of this particular task.


*   **One-Shot Learning**: As the title explains, the task only contains one training example to learn.

*   **FSL in Classification**: When were talking about FSL in classification, we usually mean N-way-K-Shot-classification. N stands for the number of classes, and K for the number of samples from each class to train on.

## Meta-Learning
Few-shot learning is a kind of meta-learning. Meta-learning is different from traditional supervised learning. Traditional supervised learning asks the model to recognize the training data and then generalize to unseen test data. Differently, meta learnings goal is to learn. In other words, in meta-learning, the goal is to find a set of parameters that can adapt as easily as possible for a new given task. In the followeing section we discuss one of the most important meta-learning platforms currently available.


### Model-Agnostic Meta-Learning
MAML, or Model-Agnostic Meta-Learning, is a model and task-agnostic algorithm for meta-learning that trains a models parameters such that a small number of gradient updates will lead to fast learning on a new task.

4.png

In this algorithm, which can be used in federated learning, each agent adapts the general set of parameters (known as theta) independently for its task using the support set. Subsequently, the gradient of the loss function using the query set (which is also labled because it is in the learning section) is computed and transfered to the root. Then, the root updates the general thera with respect to all of the recieved gradients of different agents (tasks).


[This youtube video](https://www.youtube.com/watch?v=wT45v8sIMDM&t=1286s) could help with a better understanding of MAML. Or if you are more into text, [here](https://www.instadeep.com/research/blog/model-agnostic-meta-learning-made-simple/) is a good source.

[Here](https://arxiv.org/abs/1703.03400) is the link to the original paper proposing MAML.

An easy implemntation of MAML is available on the [learn2learn](https://github.com/learnables/learn2learn) github repository. A simple experiment could be running [this python file](https://github.com/learnables/learn2learn/blob/master/examples/vision/meta_mnist.py), which executes MAML for MNIST image classification tasks.
